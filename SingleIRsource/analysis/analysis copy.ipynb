```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
hdf = h5py.File('scan_25dB_1GHz.h5', 'r')
I = np.array(hdf['i_signal'])
Q = np.array(hdf['q_signal'])
#print(I)
```

```python
I_mean = []
Q_mean = []
for i in range(len(I)):
    I_mean.append(I[i].mean())
    Q_mean.append(Q[i].mean())
```

```python
ref = 5.87045 
x = []
for i in range( -int(len(I)/2), int(len(I)/2) ):
    x.append(ref + i*0.0002) 
plt.plot(x,I_mean)
plt.xlabel('freq[GHz]')
plt.ylabel('I')
print(min(I_mean),max(I_mean))
#plt.xlim(5.865,5.875)
```

```python
plt.plot(x,Q_mean)
plt.xlabel('freq[GHz]')
plt.ylabel('Q')
print(min(Q_mean),max(Q_mean))
#plt.xlim(5.865,5.875)
```

```python
plt.scatter(Q_mean, I_mean, marker='.')
plt.xlabel('Q')
plt.ylabel('I')
```

```python
plt.plot(x,(np.array(Q_mean)**2+np.array(I_mean)**2)**0.5)
```

```python
Q_new = np.array(Q_mean) - min(Q_mean) 
I_new = np.array(I_mean) - min(I_mean) 

plt.plot(x,(np.array(Q_new)**2+np.array(I_new)**2)**0.5)
```
