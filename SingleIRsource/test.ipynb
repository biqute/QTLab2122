```python
import numpy as np
import matplotlib.pyplot as plt
import json

f = open('data1.json')
dati = json.load(f)
data = dati['Samples']['0']
x = np.linspace(0,len(data)-1,len(data))
            
plt.scatter(x, data, color = 'green', marker = '.')
plt.show()
```

```python
data2 = np.diff(data, n=1)
x = np.linspace(0,len(data2)-1,len(data2))

plt.scatter(x, data2, color = 'dodgerblue', marker = '.')
plt.show()
```

```python
n_points, i = 0, 0
max = np.std(data2[0:1000]) #al massimo si può fare due sigma
while(n_points<15): #si può aumentare n_points
    n_points = n_points + 1 if (data2[i] > max) else 0
    i += 1
start = i - 15
print(start)
```
