```python
import numpy as np
import matplotlib.pyplot as plt
import json

f = open('data1.json')
dati = json.load(f)
data = dati['Samples']['1']
x = np.linspace(0,len(data)-1,len(data))
            
plt.scatter(x[1420:1435], data[1420:1435], color = 'green', marker = '.')
plt.show()
```

```python
data2 = np.diff(data, n=1)
x = np.linspace(0,len(data2)-1,len(data2))

plt.scatter(x, data2, color = 'dodgerblue', marker = '.')
plt.show()
```

```python
n_points, i = 0, 0
std = np.std(data2[0:1000])/2 #al massimo si può fare due sigma
max = data2.argmax()
tot = max
while(data2[max])>std:
    max -= 1
tot = tot - max
while(n_points<tot): #si può aumentare n_points
    n_points = n_points + 1 if (data2[i] > std) else 0
    i += 1
start = i - tot
print(start)
```

```python
begin = start - 50 
end = start + 50
plt.scatter(x[begin:end], data[begin:end], color = 'dodgerblue', marker = '.')
plt.show()
```

```python
from scipy.signal import savgol_filter, savgol_coeffs
from scipy.interpolate import UnivariateSpline

#plt.scatter(x[begin:end], data[begin:end], color = 'green', marker = '.')
derivative_func = savgol_filter(data[begin:end], len(data[begin:end])-1, polyorder = 4, deriv = 2, delta = 1)
end=begin + len(derivative_func)
plt.scatter(x[begin:end], derivative_func, color = 'dodgerblue', marker = '.')
plt.show()
```
